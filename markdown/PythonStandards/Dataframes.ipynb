```python
import numpy as np
import pandas as pd
```

```python
# Detector coordinates
nx, ny = (256, 256) # 65536
pixel_x = np.arange(0, nx)
pixel_y = np.arange(0, ny)
pixel_x_v, pixel_y_v = np.meshgrid(pixel_x, pixel_y)

# Detector physical coordinates
nx, ny, nz = (64, 64, 16)
x = np.linspace(0, 5, nx)
y = np.linspace(0, 5, ny)
z = np.linspace(0, 2, nz)
xv, yv, zv = np.meshgrid(x, y, z)

```

```python
df = pd.DataFrame(
    {'x': xv.flatten(),
     'y': yv.flatten(),
     'z': zv.flatten(),
     'panel_id': np.ones(65536),
     'pixel_x' : pixel_x_v.flatten(),
     'pixel_y' : pixel_y_v.flatten(),
     })
```
